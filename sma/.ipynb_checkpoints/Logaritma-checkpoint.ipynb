# Logaritma

Logaritma adalah konsep terbalik dari eksponen (pangkat). 

Jika $a = b^x$, maka $\log_{b}x = a $

$\log_{b}b^{a} = a $

Di dalam beberapa buku mata pelajaran matematika, kadang $\log_{b}a$ ditulis sebagai $^b\log a $. Dalam dokumen ini kami menganut penulisan yang pertama. 

## 1. Sifat-sifat Logaritma
Untuk rumus matematika di bawah ini, $x,y,m,n,a,b,c$ dan $d$ semua positif. Kemudian diasumsikan $a\neq1$ dan $b\neq1$

### 1.1. Definisi
* $\log_{b}x = a$, mengandung makna $b^a = x$. Dalam hal ini $b$ sering disebut dengan basis logaritma.
* $\log a$ berarti $\log_{10}a$. Jika basis tidak tertulis, seperti dalam $\log a$, maka basis logaritma dianggap sebagai $10$.  Logaritma dengan basis $10$ disebut sebagai logaritma umum. 
* $\ln a$ berarti $\log_{e}a$. Logaritma natural adalah logaritma berbasis $e$, dimana nilai $e$ sekitar $2.718$. Sehingga $\ln 8 = 2.0794415$ sebab $e^{2.0794415} = 8$

### 1.2. Aturan Identitas
* $\log_{b}1 = 0$ sebab $b^{0} = 1$
* $\log_{b}b = 1$ sebab $b^{1} = b$

Catatan: $\log_{b}0$ tidak terdefinisikan, sebab tidak ada angka $x$ sedemikian sehingga $b^{x}=0$

### 1.3. Aturan Inverse (kebalikan)
* $\log_{b}b^x = x$ dan $b^{(\log_{b}x)} = x$

### 1.4. Aturan perkalian
* $\log_{b}(x.y) = \log_{b}x + \log_{b}y$ 

Jika $x = b^c$, dan $y = b^d$, maka 

$$
\begin{eqnarray}
x.y &=& b^c . b^d \\
&=& b^{c+d} \\
\log_{b}(x.y) &=& \log_{b}b^{c+d} \\
&=& c + d \\
&=& log_{b}x + \log_{b}y
\end{eqnarray}
$$

dengan jalan lain,
$$
\begin{eqnarray}
x.y &=& b^c . b^d \\
    &=& b^{\log_{b}x} . b^{\log_{b}y} \\
    &=& b^{\log_{b}x + \log_{b}y} \\
\log_{b}(x.y) &=& \log_{b}x + \log_{b}y
\end{eqnarray}
$$

#### Hati-hati:
$\log_{b}(x + y) \neq \log_{b}x + \log_{b}y$


### 1.5. Aturan pembagian
* $\log_{b}\left(\frac{x}{y}\right) = \log_{b}x - \log_{b}y$

Jika $x = b^c$, dan $y = b^d$, maka 
$$
\begin{eqnarray}
x:y &=& b^c : b^d \\
    &=& b^{c-d} \\
\log_{b}(\frac{x}{y}) &=& \log_{b}b^{c-d} \\
                      &=& c - d \\ 
                      &=& log_{b}x - \log_{b}y
\end{eqnarray}
$$

#### Hati-hati:
$\log_{b}(x - y) \neq \log_{b}x - \log_{b}y$

### 1.6. Aturan eksponen/pangkat
* $\log_{b}x^n = n.\log_{b}x$

Jika $x = b^c$, maka 
$$
\begin{eqnarray}
x^n &=& (b^c)^n \\
&=& b^{c.n} \\
\log_{b}x^n &=& \log_{b}b^{c.n} \\
            &=& c . n \\
            &=& n . \log_{b}x 
\end{eqnarray}
$$

Dengan aturan pangkat ini, maka kita bisa buktikan aturan pembagian di atas dengan cara pembuktian lain sebagai berikut,

$$
\begin{eqnarray}
\log_{b}\left(\frac{x}{y}\right) &=& \log_{b}(x.y^{-1}) \\
&=& \log_{b}x + \log_{b}y^{-1} \\
&=& \log_{b}x - \log_{b}y 
\end{eqnarray}
$$


### 1.7. Aturan perubahan basis
* $\log_{b}x = \frac{\log_{d}x}{\log_{d}b} $

Pembuktian:
Jika $x = b^c$ maka $\log_{b}x = c$ sehingga jika dioperasikan $\log_{d}$ pada kedua ruas

$$
\begin{eqnarray}
log_{d}x &=& \log_{d}b^c \\
log_{d}x &=& c . \log_{d}b \\
c &=& \frac{log_{d}x}{\log_{d}b} \\
\log_{b}x &=& \frac{log_{d}x}{\log_{d}b}
\end{eqnarray}
$$


### Turunan dari aturan-aturan 
Berdasarkan aturan-aturan di atas kita dapat menurunkan aturan-aturan lain sebagai berikut:
* $\log_{b}(\sqrt[y]{x}) = \frac{\log_{b}x}{y}$

Sebab, 
$$
\begin{eqnarray}
\log_{b}(\sqrt[y]{x}) &=& \log_{b}(x^{\frac{1}{y}})\\
                      &=& \frac{1}{y} . \log_{b}x  \\
                      &=& \frac{\log_{b}x}{y} \\
\end{eqnarray}
$$

* $x^{\log_{b}y} = y^{\log_{b}x}$
